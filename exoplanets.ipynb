```python
import matplotlib.pyplot as plt 
import numpy as np
from collections import Counter
import kagglehub
import os 
import pandas as pd

#download material 
path=kagglehub.dataset_download('adityamishraml/nasaexoplanets')
print('Path is:',path)

#csv path
file_path=os.path.join(path,'cleaned_5250.csv')
df=pd.read_csv(file_path)

planet_type=df["planet_type"]




counts=Counter(planet_type)

plt.bar(counts.keys(),counts.values())
plt.xlabel("Planet Type")
plt.ylabel("Count")
plt.title("Planet Frequency")
plt.show()
```

```python
import kagglehub
import os
import pandas as pd
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt 

path=kagglehub.dataset_download('adityamishraml/nasaexoplanets')
print("Path is:",path)

file_path=os.path.join(path,'cleaned_5250.csv')
df=pd.read_csv(file_path)

mass_wrt=df["mass_wrt"]

counts=Counter(mass_wrt)

plt.bar(counts.keys(),counts.values())
plt.xlabel("mass_wrt")
plt.ylabel("Count")
plt.title("With respect to:")
plt.show()
```

```python
import matplotlib.pyplot as plt
from collections import Counter

# Drop missing values
mass_wrt_clean = df["mass_wrt"].dropna()

# Count occurrences
counts = Counter(mass_wrt_clean)

# Plot
plt.bar(counts.keys(), counts.values(), color='skyblue')
plt.xlabel("Mass Reference")
plt.ylabel("Count")
plt.title("Mass Measured With Respect To Earth")
plt.xticks(rotation=45)  # Rotate labels if needed
plt.show()
```
